```python
import requests
from bs4 import BeautifulSoup

url = "https://analyticsindiamag.com/8-companies-offering-the-highest-salaries-to-data-scientists-in-india/"
response = requests.get(url)
html_content = response.text

# Parse the HTML content using BeautifulSoup
soup = BeautifulSoup(html_content, 'html.parser')

# Find all h3 tags containing company names
company_names = soup.find_all("h3")

# Check if any h3 tags are found
if company_names:
    # Iterate over each company name
    for company_name in company_names:
        # Extract company name
        company = company_name.text.strip()
        print("Company:", company)

        # Find the parent div containing all the information about the company
        company_info = company_name.find_next('div')

        # Extract company profile if available
        company_profile_tag = company_info.find("p")
        if company_profile_tag:
            company_profile = company_profile_tag.text.strip()
            print("Company Profile:", company_profile)
        else:
            print("Company Profile: Not available")

        # Extract job profile if available
        job_profile_tag = company_info.find("p").find_next("p")
        if job_profiles_tag:
            job_profile = job_profile_tag.text.strip()
            print("Job Profile:", job_profile)
        else:
            print("Job Profile: Not available")

        # Extract location if available
        location_tag = company_info.find("p").find_next("p").find_next("p")
        if location_tag:
            location = location_tag.text.strip()
            print("Location:", location)
        else:
            print("Location: Not available")

        # Extract duration if available
        duration_tag = company_info.find("p").find_next("p").find_next("p").find_next("p")
        if duration_tag:
            duration = duration_tag.text.strip()
            print("Duration:", duration)
        else:
            print("Duration: Not available")

        # Extract stipend if available
        stipend_tag = company_info.find("p").find_next("p").find_next("p").find_next("p").find_next("p")
        if stipend_tag:
            stipend = stipend_tag.text.strip()
            print("Stipend:", stipend)
        else:
            print("Stipend: Not available")

        # Extract requirements if available
        requirements_tag = company_info.find("ul")
        if requirements_tag:
            requirements = requirements_tag.text.strip()
            print("Requirements:", requirements)
        else:
            print("Requirements: Not available")

        print("\n")
else:
    print("No company names found in the HTML content.")
```

```python
import requests
from bs4 import BeautifulSoup
import csv
# Make a request
page = requests.get(
    "https://codedamn-classrooms.github.io/webscraper-python-codedamn-classroom-website/")
soup = BeautifulSoup(page.content, 'html.parser')

# Create top_items as empty list
all_products = []

# Extract and store in top_items according to instructions on the left
products = soup.select('div.thumbnail')
for product in products:
    name = product.select('h4 > a')[0].text.strip()
    #> is the child combinator in CSS selectors. It selects elements that are direct children of a specified parent element
    description = product.select('p.description')[0].text.strip()
    # targets all <p> elements with the class description
    price = product.select('h4.price')[0].text.strip()
    # targets all <h4> elements with the class price
    reviews = product.select('div.ratings')[0].text.strip()
    # in rating
    image = product.select('img')[0].get('src')

    all_products.append({
        "name": name,
        "description": description,
        "price": price,
        "reviews": reviews,
        "image": image
    })


keys = all_products[0].keys()

with open('products.csv', 'w', newline='') as output_file:
    dict_writer = csv.DictWriter(output_file, keys)
    dict_writer.writeheader()
    dict_writer.writerows(all_products)
```
